```python
import numpy as np  # Import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Generate synthetic 3D data
np.random.seed(42)
X = np.random.rand(10, 3) * 10  # 10 points with random X, Y, Z values scaled by 10

# Apply K-Means with K=2
kmeans_2 = KMeans(n_clusters=2)
kmeans_2.fit(X)
clusters = kmeans_2.predict(X)

# Print clusters and centroids
print("Clusters:", clusters)
print("Cluster Centers:\n", kmeans_2.cluster_centers_)

# Standardize the data for better visualization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Visualize the clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='rainbow')
plt.scatter(kmeans_2.cluster_centers_[:, 0], kmeans_2.cluster_centers_[:, 1], color='black', marker='X')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('K-Means Clustering on Synthetic 3D Data (K=2)')
plt.show()
```
