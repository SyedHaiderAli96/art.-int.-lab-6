```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic dataset
np.random.seed(42)
age = np.random.randint(18, 65, size=100)  # Random ages between 18 and 65
income = np.random.randint(20000, 100000, size=100)  # Random income between 20k and 100k

# Create DataFrame
df = pd.DataFrame({'Age': age, 'Income($)': income})

# Apply K-Means with K=3
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(df[['Age', 'Income($)']])

# Print cluster centers
print("Cluster Centers:", kmeans.cluster_centers_)

# Plot the clusters
plt.scatter(df['Age'], df['Income($)'], c=df['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='X')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.title('K-Means Clustering with Synthetic Data (K=3)')
plt.show()
```
