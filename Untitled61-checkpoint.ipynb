```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for 10 individuals (Age and Income)
np.random.seed(42)  # For reproducibility
ages = np.random.randint(18, 60, size=10)  # Random ages between 18 and 60
incomes = np.random.randint(30000, 80000, size=10)  # Random incomes between 30k and 80k

# Create a DataFrame
df = pd.DataFrame({
    'Age': ages,
    'Income($)': incomes
})

# Preprocessing: MinMax Scaling to normalize Age and Income
scaler = MinMaxScaler()
df[['Age', 'Income($)']] = scaler.fit_transform(df[['Age', 'Income($)']])

# Apply K-Means with K=3
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(df[['Age', 'Income($)']])

# Print cluster centers
print("Cluster centers:", kmeans.cluster_centers_)

# Plotting clusters
plt.scatter(df['Age'], df['Income($)'], c=df['Cluster'], cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', marker='X')
plt.xlabel('Age')
plt.ylabel('Income ($)')
plt.title('K-Means Clustering (K=3)')
plt.show()
```
